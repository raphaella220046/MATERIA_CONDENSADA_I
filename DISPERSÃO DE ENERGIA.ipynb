### Poliacetileno

O poliacetileno $(CH)_x$ é uma molécula orgânica conjugada de geometria linear. O polímero apresenta a estrutura CIS, porém transforma-se na forma TRANS por processo de isomerização térmica que é mais estável. [3]

O poliacetileno é um mal condutor elétrico, porém o baixo valor do potencial de ionização e sua alta afinidade eletrônica indicam que ele pode ser dopado tanto por receptores de elétrons (dopagem tipo p) quanto por doadores de elétrons (dopagem tipo n), e por isso ele é estudado em diversas literaturas envolvendo diferentes dopantes, a fim de melhotar a sua condutividade elétrica. ### Teorema de Bloch

O teorema de Bloch é um princípio fundamental na teoria de bandas para descrever a estrutura eletrônica em materiais periódicos.

As funções de onda de Bloch são soluções das equações de Schrödinger para elétrons em uma rede periódica. Elas são definidas como:

$\psi(k, r) = u(k, r) e^{(ikr)}$

Em que:

$\psi(k, r)$ é a função de onda de Bloch.

$k$ é o vetor de onda na rede recíproca.

$r$ é a posição na rede real.

$u(k, r)$ é uma função periódica no espaço real, com a mesma periodicidade da rede real.

Essas funções de onda são moduladas por um fator exponencial complexo que depende do vetor de onda k

$$\psi(k, r) = \sum(n) [c_n \phi_{n(r)} e^{(ikr)}]$$

Onde:

$\psi(k, r)$ é a função de onda de Bloch.

$c_n$ são os coeficientes de expansão.

$\phi_{n(r)}$ são as funções de onda dos átomos de carbono individuais.

$e^{(ikr)}$ é o fator de Bloch.

#### O método de Huckel

É uma técnica teórica simplificada usada para calcular as energias eletrônicas e as propriedades de sistemas conjugados de elétrons em moléculas planares orgânicas. Esse método é particularmente útil para estudar sistemas de pi-conjugados, nos quais os elétrons podem delocalizar-se ao longo de uma cadeia de átomos. A principal simplificação do método de Huckel é a descrição dos orbitais moleculares $\pi$ em termos de orbitais atômicos p. No método de Huckel também é feita a simplificação da integral de overlap $S=0$.

A matriz de representação da Hamiltoniana $[H]$ pode ser escrita em termos da energia do orbital de cada átomo. A diagnoal principal representa as energias dos orbitais $\pi$ dos átomos de carbonos individuais $\alpha$, enquanto  como  sítios vizinhos a cada carbono da célula periódica se alternam entre $\beta$ e $\gamma$.

```python
βαγ = symbols('βαγ', negative = True)
γαβ = symbols('γαβ', negative = True)

H = Matrix([
[βαγ, 0, 0, 0, 0, 0],
[0, γαβ, 0, 0, 0, 0],
[0, 0, βαγ, 0, 0, 0],
[0, 0, 0, γαβ, 0, 0],
[0, 0, 0, 0, βαγ, 0],
[0, 0, 0, 0, 0, γαβ]])

display(H)
```

```python
H.eigenvects()
```

```python
det_H = H.det()
display(det_H)
```

A dispersão de energia é obtida

Para $\phi_{Aj}$:

$$\large \beta \phi_{(Bj-1)} + \alpha \phi_{(Aj)} + \gamma \phi_{(Bj)} = E \phi_{Aj}$$

Para $\phi_{Bj}$:
$$\large \gamma \phi_{(Aj)} + \alpha \phi_{(Bj)} + \beta \phi_{(Aj+1)} = E \phi_{Bj}$$

Aplicação do teorema de Bloch com fator de fase $(ika)$ para as ligações entre sítios

Para $\phi_{Aj}$:
$$\large \beta e^{-ika} \phi_{(Bj)} + \alpha \phi_{(Aj)} + \gamma \phi_{(Bj)} = E \phi_{Aj}$$

Para $\phi_{Bj}$:
$$\large \gamma \phi_{(Aj)} + \alpha \phi_{(Bj)} + \beta e^{-ika} \beta \phi_{(Aj)} = E \phi_{Bj}$$

Essa matriz pode ser reescrita nos termos de $$H\psi = E\psi$$ da seguinte forma:

```python
h11, h12, h21, h22 = symbols('α γ+βe^(-ika) γ+βe^(ika) α', negative = True)

h = Matrix([
[h11, h12],
[h21, h22]])

h11, h21 = symbols('𝜑_A 𝜑_B', negative = True)

𝜓AB = Matrix([
[h11],
[h21]])

display(h)
display(𝜓AB)
print("= E")
display(𝜓AB)
```

Da equação secular, tem-se $$det[\alpha-\lambda I]=0$$

Que resulta na seguinte equação do segundo grau

$$(\alpha - \lambda)^2 $$

Com duas raízes possíveis

```python
h_𝜆11, h_𝜆12, h_𝜆21, h_𝜆22 = symbols('(α-𝜆) (γ+βe^{(-ika)}) (γ+βe^{(ika)}) (α-𝜆)', negative = True)

h_𝜆 = Matrix([
[h_𝜆11, h_𝜆12],
[h_𝜆21, h_𝜆22]])

display(h_𝜆)
```

```python
det_h = h_𝜆.det()
display(det_h)
```

Para encontrar os níveis de energia, é preciso isolar $\lambda$, sendo assim

$$(\alpha - \lambda)^2 - [\gamma^2 + \boxed{\gamma \beta e^{(ika)} + \gamma \beta e^{(-ika)}} + \beta^2]$$

Colocando em evidência o termo $\gamma \beta e^{(ika)} + \gamma \beta e^{(-ika)} \implies \gamma \beta (e^{-ika}+e^{ika})$, este pode ser escrito em termos do cosseno, como $2 cos(ka)$, então $$(\alpha - \lambda)^2 - [\gamma^2 + \beta^2 + 2 \gamma \beta cos(ka)]=0$$

Logo, as duas raízes possíveis são

$$\lambda = \pm \alpha \sqrt{\gamma^2 + \beta^2 + 2\gamma \beta cos(ka)}$$

Que representam os autovalores de energia para esse sistema.

#### Estrutura de bandas

A estrutura de bandas pode ser determinada quando $E_-(k)$ e $E_+(k)$

```python
b=10
y=5
a=1
alpha=1
k=np.linspace(-10,10,1000)

def energia_positiva(b,y,k,a):
    return alpha+((y**2)+(b**2)+(2*y*b*np.cos(k*a)))**(1/2)

def energia_negativa(b,y,k,a):
    return alpha-((y**2)+(b**2)+(2*y*b*np.cos(k*a)))**(1/2)
```

```python
plt.rcParams['font.family'] = 'Verdana'
plt.figure(figsize = (15, 10))
plt.plot(k, energia_positiva(b,y,k,a), label = '$\pi$* band', linewidth=5, linestyle='-', color='mediumturquoise')
plt.plot(k, energia_negativa(b,y,k,a), label = '$\pi$ band', linewidth=5, linestyle='-', color='lightpink')
plt.xlabel('k (in units of $\pi$/a)', size=30)
plt.ylabel('Energy (eV)', size=30)
plt.title('Bandstructure', size=40)
plt.tick_params(labelsize=20)
plt.rcParams['legend.fontsize'] = 20
#plt.xlim(-1,150)
#plt.ylim(-1,5)
plt.grid()
plt.legend()
#plt.savefig('...')
plt.show()
```

#### Interpretação gráfica

O gráfico da estrutura de banda do poliacetileno apresenta a energia em função do vetor de onda do elétron $\vec{k}$, permitindo analisar a distribuição dos níveis de energia ao longo da cadeia e o comportamento dos elétrons em relação à esses níveis.

Observa-se um comportamento periódigo ao longo da cadeia, em que as bandas $\pi$ e $\pi*$ são simétricas e invertidas, e correspondem aos orbitais ligante (banda de valência) e antiligante (banda de condução), respectivamente. Essa periodicidade ocorre tanto no espaço simétrico quanto no espaço recíproco.

A medida que se desloca para a direita a banda $\pi$ aumenta em energia enquanto a banda $\pi *$ decresce. Entre as bandas é observada a abertura do band gap que aumenta ou diminui ao longo da cadeia. Esse fenômeno ocorre devido a distorção ou instabilidade de Peierls para sistemas cristalinos unidimensionais com ligações fortes, induzindo um gap de 1,5 eV e a alternância do comprimento da ligação representa a maior contribuição para a divisão energética do nível HOMO e LUMO. [9]

O topo da banda de valência em $k=\frac{\pi}{a}$ tem uma função de onda que constitui o nível HOMO (Highest Occupied Molecular Orbital) da molécula. Existem fatores de ligação onde no estado fundamental existem ligações duplas, e fatores antiligantes onde no estado fundamental existem ligações simples. Nessa mesma região há uma situação de antiligação dentro da célula unitária e uma situação de ligação entre células unitárias. A ligação e anti-ligação LUMO (Lowest Unoccupied Molecular Orbital) são opostas às do HOMO. Sendo assim, há antiligação onde existe um fator de ligação duplo e ligação onde existe um fator de ligação do estado fundamental simples. [8]

#### Análise da dispersão de energia para $\beta>\gamma$ e $\beta<\gamma$

Tanto no cenário em que $\beta>\gamma$ ou $\beta<\gamma$, os elétrons em um estado fundamental ocuparão os orbitais de energia mais baixos que se alternam entre $\beta$ e $\gamma$. Na análise gráfica não foi notada diferença na estrutura de banda ao alternar entre os sítios de ligação, ou seja, ligação simples e dupla, de menor ou maior energia. Isso ocorre devido aos pontos de simetria da molécula, não interferindo na mobilidade eletrônica, no alargamento ou estreitamento das bandas de energia (maior/menor dispersão de energia).

Quando a $\beta = \gamma$ o gap de energia mencionado anteriormente desaparece nos pontos onde as bandas se tocam, não havendo nessa região a transição proibida para os elétrons. Isso significa que a banda de valência e a banda de condução se encontram, e a transição eletrônica entre elas é permitida, fazendo com que os elétrons alcancem níveis de energia mais altos. No material semicondutor, isso significa que quando exposto a uma diferença de potencial, ele pode conduzir eletricidade. Isso tem relação com o ajuste por dopagem para tornar o poliacetileno um polímer condutor, como abordado antes.

Portanto, a dispersão de energia para o poliacetileno aplicando o modelo de Huckel com hopping alternado entre $\beta$ e $\gamma$ não tem a estrutura de bandas alterada por suas energias relativas, o que não afeta as propriedades eletrônicas e de transporte do poliacetileno entre os dois sítios.
