### Análise dimensional da densidade de elétrons $g(\epsilon)$

$$\large g(\epsilon) = \frac{V}{2 \pi^2} \left( \frac{2m}{\hbar^{2}} \right) ^\frac{3}{2} \epsilon^{\frac{1}{2}}$$

$\large V \rightarrow m^3$

$\large 2 \pi^2 \rightarrow $ adimensional

$\large m \rightarrow kg$

$\large \hbar \rightarrow \frac{m^2 kg}{s}$

$\large \epsilon \rightarrow J \rightarrow \frac{kg m^2}{s^2}$

$$\large g(\epsilon) = m^3 \cdot \left( \frac{kg}{\left(\frac{m^2 kg}{s}\right)^2} \right)^{\frac{3}{2}} \cdot \left( \frac{kg m^2}{s^2} \right) ^{\frac{1}{2}}$$

$$\large m^{3} \cdot m^{-6} \cdot m^{1} \cdot kg^\frac{3}{2} \cdot kg^{-3} \cdot kg^{\frac{1}{2}} \cdot s^{3} \cdot s^{-1}$$

$$\large g(\epsilon) = m^{-2} \cdot kg^{-1} \cdot s^{2} \rightarrow \frac{1}{J}$$

### Densidade de elétrons para um gás 3D em função da energia $g(e)\cdot f(e)$

$$\large f(\epsilon) = \frac{1}{e^{\frac{\epsilon - \mu}{K_B T}}+1}$$

```python
V = 1 #1000*10**(-30) # m³
m = 1 #9.109*10**(-31) #kg
hbar = 1 #6.62607015*10**(-34) #m²kg/s

mi = 5 #6.4*10**(-19) # Potencial químico
K_B = 1 #1.380649*10**(-23) #m²kg/s²K
T = 0.01 # Temperatura em K

e = np.linspace(0, 10, 100) # J

# Considerando hbar = 1

def g(e):
    #return V/(2*np.pi**2)*(2*m/hbar**2)**(3/2)*e**(1/2)
    return e**(1/2)

# Considerando K_B = 1

def f(e, T, mi):
    #return 1/(np.exp((e-mi)/(K_B*T))+1)
    return 1/(np.exp((e-mi)/(T))+1)

# Representação gráfica

plt.rcParams['font.family'] = 'Verdana'
plt.figure(figsize = (15, 10))
plt.plot(e, g(e), label = '$g(\epsilon)$', linewidth=5, linestyle='-', color='mediumturquoise')
plt.plot(e, f(e, T, mi), label = '$f(\epsilon)$ a T = 0', linewidth=5, linestyle='-', color='lightpink')
plt.plot(e, f(e, 0.1, mi), label = '$f(\epsilon)$ a T = 0,1', linewidth=5, linestyle='-', color='lightsalmon')
plt.plot(e, f(e, 1, mi), label = '$f(\epsilon)$ a T = 1', linewidth=5, linestyle='-', color='lightcoral')
plt.plot(e, f(e, 10, mi), label = '$f(\epsilon)$ a T = 10', linewidth=5, linestyle='-', color='indianred')
plt.plot(e, g(e)*f(e, T, mi), label = 'Densidade eletrônica = $g(\epsilon) \cdot f(\epsilon)$', linewidth=5, color='violet')
plt.xlabel('Energia ($\epsilon$)', size=30)
plt.ylabel('Probabilidade de ocupação', size=30)
plt.title('Função de distribuição de Fermi-Dirac', size=40)
plt.tick_params(labelsize=20)
plt.rcParams['legend.fontsize'] = 20
#plt.xlim(-1,150)
#plt.ylim(-1,5)
plt.grid()
plt.legend()
#plt.savefig('...')
plt.show()
```

#### DISCUSSÃO

A função de distribuição de Fermi-Dirac descreve a probabilidade de ocupação de diferentes estados de energia disponíveis por partículas fermiônicas (que apresentam spin semi-inteiro, como elétrons) em um sistema quântico em equilíbrio térmico.

O gráfico apresenta a função de distribuição de Fermi-Dirac para diferentes temperaturas. Observa-se que a função varia entre 0 e 1, representando a probabilidade de um estado de energia ser ocupado por elétrons. Ao alterar a temperatura, ocorre a transferência de uma parte dos elétrons em uma região de energia da ordem de $k_B T$ em relação ao potencial químico $\mu$ para os estados acima de $\mu$. Em temperaturas próximas a $T=0$ a função se aproxima de um "degrau", onde os estados de energia abaixo da energia de Fermi são altamente ocupados (probabilidade próxima de 1) e os estados acima da energia de Fermi são pouco ocupados (probabilidade próxima de 0). À medida que a temperatura aumenta, há uma maior mobilidade eletrônica, aumentando com isso a densidade eletrônica, devido ao ganho de energia cinética dos elétrons. Nessa situação, a função de distribuição apresenta uma transição suave. Assim, é demonstrado que a probabilidade de ocupação de diferentes estados de energia muda com a temperatura.

#### Pressão em um gás de eletrons livres à T=0

Na temperatura zero, a distribuição de Fermi-Dirac é expressa por

$$ f(\epsilon)=
\begin{align*}
1, \epsilon<\mu, \\
1/2, \epsilon=\mu, \\
0, \epsilon>\mu, \\
\end{align*}
$$

O que significa que orbitais com energias menores que $\mu$ estão ocupados e aqueles com energia maiores do que $\mu$ estão vazios. O número médio de elétrons pode ser obtido ao integrar $N = \int_0^\infty f(\epsilon)G(\epsilon)d\epsilon \implies N=\int_0^\mu G(\epsilon)d\epsilon=\eta (\mu)$, ou seja, o número de orbitais $\eta$ menores do que $\mu$.

O número de orbitais entre zero e $\epsilon$ é determinado ao calcular o número de vetores de onda $k$, tais que $\epsilon_{\vec{k}}=\frac{\hbar^2}{2m}(k_x^2+k_y^2+k_z^2)\leq \epsilon$ e multiplicando o resultado por 2, pois cada vetor de onda corresponde a dois orbitais. Ao associar cada possível vetor em um ponto no espaço tridimensional $(k_x, k_y, k_z)$, o número de vetores de onda é igual ao número de cubos elementares de volume $(\frac{2 \pi}{L}^3)=\frac{8 \pi^3}{V}$ em uma esfera de raio $\sqrt{\frac{2m\epsilon}{\hbar^2}}$.

Sendo assim, $$\eta(\epsilon)=2\frac{\frac{4}{3}\pi \left( \frac{2m\epsilon}{\hbar^2} \right)^\frac{3}{2}}{\frac{8\pi^3}{V}}=\frac{V}{3\pi^2}\left( \frac{2m}{\hbar^2} \right)^\frac{3}{2} \epsilon^\frac{3}{2}.$$

Ao derivar $\large d\frac{\eta(\epsilon)}{d(\epsilon)}$ é obtida a densidade de orbitais $\large g(\epsilon) = \frac{V}{2 \pi^2} \left( \frac{2m}{\hbar^{2}} \right) ^\frac{3}{2} \epsilon^{\frac{1}{2}}$.

A partir da equação $\eta(\epsilon)$ para é possível obter o número de elétrons como sendo $$\large N =\frac{V}{3 \pi^2} \left( \frac{2m}{\hbar^2} \right)^\frac{3}{2} \mu^\frac{3}{2}.$$

E a partir de $\large U=\int_0^\infty \epsilon f(\epsilon) G(\epsilon) d\epsilon$ a energia $E$ é dada por $$\large E =\frac{V}{5 \pi^2} \left( \frac{2m}{\hbar^2} \right)^\frac{3}{2} \mu^\frac{5}{2}$$

A razão entre $\frac{E}{N} = \frac{3}{5} \mu$. Na temperatura zero a energia de Fermi coincide com o potencial químico $\epsilon_F = \mu$, logo $$\frac{E}{N}=\frac{3}{5} \epsilon_F \implies E=\frac{3}{5}N \epsilon_F,$$ sendo que $\large \epsilon_F=\frac{\hbar^2}{2m}\left( 3\pi^2\frac{N}{V} \right)^\frac{2}{3}$

A partir dessa equação é possível calcular a pressão do gás de elétrons livres, escrevendo a energia total como uma função explícita de $N$ e $V$, da seguinte maneira

$$\large E=2 \sum_{k<k_F} \epsilon(k)=2\frac{V}{(2\pi)^3} \int_{k<k_F}dk\frac{\hbar^2k^2}{2m}=\frac{V}{4\pi^3}\int_0^{k_F}dk(4\pi k^2)\frac{\hbar^2k^2}{2m}=\frac{\hbar^2 k_F^5V}{10\pi^2m}$$

$$\large P = -\left(\frac{\partial E}{\partial V}\right)_N = \frac{2}{5}N\frac{\hbar^2}{2m}(3\pi^2N)^{\frac{2}{3}}V^{\frac{-5}{3}}=\frac{2}{3}\frac{E}{V}$$

Ao substituir $E=\frac{3}{5}N \epsilon_F$ em $P=\frac{2}{3}\frac{E}{V}$, tem-se $PV=\frac{2}{5}N\epsilon_F$

**Discussão**

A pressão em um gás de elétrons livres a uma temperatura de zero absoluto ($T = 0 K$) pode ser encontrada usando o modelo do elétron livre e os princípios da mecânica quântica. No entanto, o modelo do elétron livre é uma simplificação teórica que assume que os elétrons estão completamente descorrelacionados e livres para se moverem em um potencial contínuo, o que pode não ser completamente preciso em todos os casos.

A pressão $P$ em um gás de elétrons livres é dada pela expressão:

$P = \frac{2}{3} \cdot \frac{E}{V}$

Onde:
- $E$ é a energia total dos elétrons no sistema;
- $V$ é o volume ocupado pelo gás de elétrons livres.

Nesse contexto, o gás de elétrons livres é frequentemente associado a metais, onde os elétrons estão livres para se moverem no interior da rede cristalina.

A expressão acima surge da relação entre a energia cinética média dos elétrons e a pressão exercida por eles nas paredes de contorno. Ela implica que a pressão é diretamente proporcional à densidade de energia dos elétrons e, por sua vez, à densidade de elétrons no sistema.

O resultado evidencia que a pressão de um gás de elétrons livres a temperatura zero não é nula. Isso acontece porque os elétrons continuam a ter uma energia cinética associada ao seu movimento quântico. Portanto, a pressão não desaparece completamente nessa condição.

#### Predições positivas e negativas do modelo do elétron livre

O modelo do elétron livre é uma simplificação teórica que descreve o comportamento dos elétrons em um metal como se eles estivessem livres para se mover em um potencial contínuo, desconsiderando as interações entre elétrons e com o núcleo. Mesmo que esse modelo tenha algumas limitações, ele ainda é útil para explicar muitos comportamentos observados em metais, mas também apresenta algumas predições positivas e negativas, especialmente quando comparado à realidade mais complexa dos materiais.

**Predições Positivas do Modelo do Elétron Livre:**

1. **Condutividade Elétrica:** O modelo do elétron livre pode explicar a alta condutividade elétrica observada em metais, uma vez que os elétrons "livres" têm grande mobilidade em um potencial contínuo.

2. **Condutividade Térmica:** O modelo também pode explicar a alta condutividade térmica dos metais, uma vez que os elétrons livres podem transferir energia térmica rapidamente.

3. **Efeito Hall:** O modelo pode explicar o Efeito Hall, que é a geração de uma diferença de potencial transversal a uma corrente elétrica e um campo magnético aplicado, indicando que os elétrons têm carga negativa.

4. **Propriedades Ópticas Simplificadas:** Em algumas situações, o modelo pode fornecer uma explicação qualitativa para a absorção de luz em metais, embora não leve em consideração as bandas de energia reais.

**Predições Negativas do Modelo do Elétron Livre:**

1. **Capacidade Calorífica e Espectro de Calor Específico:** O modelo do elétron livre prediz uma capacidade calorífica eletrônica linear com a temperatura, o que difere significativamente da capacidade calorífica observada em metais.

2. **Magnetismo:** O modelo do elétron livre não pode explicar fenômenos magnéticos em metais, como a ferromagnetismo ou o paramagnetismo.

3. **Ligação Química e Propriedades Não-Metálicas:** O modelo não considera a ligação química entre átomos que é essencial para explicar as propriedades de materiais não-metálicos e moleculares.

4. **Efeitos de Correlação Eletrônica:** O modelo não leva em conta as interações de correlação entre elétrons, que são cruciais para entender fenômenos como supercondutividade e semicondutores intrinsecamente.

De acordo com ASHCROFT e MERMIN (2011) as deficiências mais notáveis do modelo se devem ao uso da mecânica estatística clássica na descrição dos elétrons de condução. Como consequência, os campos termoelétricos e as capacidades caloríficas previstos eram centenas de vezes maiores, mesmo a temperatura ambiente. A aplicação da estatística de Fermi-Dirac feita por Sommerfeld nos elétrons de condução elimiou esta classe de dificuldades, mantendo as outras suposições básicas do modelo do elétron livre. No entanto, algumas inadequações são observadas como:

- **Os coeficientes de transporte do elétron livre:** O coeficiente de Hall independente da temperatura, do tempo de relaxação ou da força do campo magnético, a magnetorresistência em que a resistência de um fio perpendicular a um campo magnético uniforme independe da força do campo, o sinal do campo termoelétrico nem sempre é o mesmo expresso pelo modelo, a lei de Wiedemann-Franz não se aplica a todas as temperaturas, a falta de explicação sobre a dependência da temperatura na condutividade DC, falha ao explicar a dependência da frequência em relação às propriedades ópticas dos metais para além da constante dielétrica do elétron livre.

- **Previsões da termodinâmica estatística:** o termo linear no calor específico não é tão bem explicado para os metais nobres ou metais de transição, o termo cúbico no calor específico não é justificado, a compressibilidade dos metais que considera o cálculo de módulos bulk, mas não de íons e interações elétron-elétron para uma estimativa mais acurada da equação do estado de um metal.

Outras inadequações são quanto a falta de explicação acerca dos elétrons de valência que se tornam elétrons de condução, enquanto os demais permanecem ligados aos íons, e falta de justificativa sobre os elementos que não são metais.

Portanto, o modelo do elétron livre é uma aproximação útil para entender algumas propriedades elétricas e térmicas de metais, mas não pode explicar completamente as propriedades de todos os materiais, especialmente quando interações eletrônicas, efeitos magnéticos, ligação química e correlações eletrônicas são relevantes. Os elementos faltantes no modelo incluem a consideração das bandas de energia, efeitos de correlação eletrônica, estrutura de bandas real e interações de spin que são essenciais para descrever uma gama mais ampla de materiais e comportamentos.
